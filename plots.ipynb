```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import jn_zeros, j0

u04 = jn_zeros(0, 4)[3]

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

x = np.linspace(-1, 1, 400)
y = np.linspace(-1, 1, 400)
X, Y = np.meshgrid(x, y)
R = np.sqrt(X**2 + Y**2)

Z = j0(u04 * R)
Z[R > 1] = np.nan

ax.plot_surface(X, Y, Z, cmap='plasma', edgecolor='none', antialiased=True)

ax.set_xlabel('x', labelpad=10)
ax.set_ylabel('y', labelpad=10)
ax.set_zlabel(r'$J_0(u_{04} r)$', labelpad=10)
ax.set_title(r'$J_0(u_{04} r)$ on unit disk')

# Fewer ticks to avoid overlap
ax.set_xticks([-1, 0, 1])
ax.set_yticks([-1, 0, 1])
ax.set_zticks([-0.4, 0, 0.4, 0.8])

ax.view_init(elev=25, azim=45)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(figsize=(6, 5))

# Axes
ax.annotate('', xy=(3, 0), xytext=(-2, 0), arrowprops=dict(arrowstyle='->', color='black', lw=1))
ax.annotate('', xy=(0, 2.5), xytext=(0, -1.5), arrowprops=dict(arrowstyle='->', color='black', lw=1))

# Point and radius
theta = np.radians(125)
r = 1.8
x, y = r * np.cos(theta), r * np.sin(theta)

ax.plot([0, x], [0, y], 'k-', lw=1.5)  # radius line
ax.plot(x, y, 'ro', markersize=8)       # red point

# Dashed extension
gap = 0.15
ext = 0.5
x_start, y_start = x + gap * np.cos(theta), y + gap * np.sin(theta)
x_ext, y_ext = x_start + ext * np.cos(theta), y_start + ext * np.sin(theta)
ax.plot([x_start, x_ext], [y_start, y_ext], 'k--', lw=1.5, dashes=(5, 3))

# Angle arc
arc_theta = np.linspace(0, theta, 50)
arc_r = 0.7
ax.plot(arc_r * np.cos(arc_theta), arc_r * np.sin(arc_theta), 'k-', lw=1)

# Labels
ax.text(0.85, 0.65, r'$\theta$', fontsize=16)
ax.text(x/2 + 0.1, y/2 + 0.1, r'$r$', fontsize=14)

ax.set_xlim(-2, 3)
ax.set_ylim(-1.5, 2.5)
ax.set_aspect('equal')
ax.axis('off')
plt.tight_layout()
plt.savefig('polar_coords.png', dpi=150, bbox_inches='tight')
plt.show()
```
